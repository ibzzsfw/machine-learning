# Normalize data option

- Add an option to `Naive Bayes and KNN Iris and Cancer.ipynb` to ask the user whether to normalize the dataset by converting each feature into a **Z-distribution** by making **mean = 0**, and **standard deviation = 1**. 
- For this problem, compare the accuracy results for the breast cancer dataset on the sklearn's GaussianNB classifier using normalized vs. unnormalized data.

```python
from sklearn.datasets import load_breast_cancer
the_data = load_breast_cancer() #get the data from sklearn
```

```python
print("\nThe Entire Dataset \n", the_data)
```

```python
# the_data is an extension of the dictionary data type into a sklearn type called bunch.  The Iris data is a dictionary data type.
# The dicitionay has the following key -> values:
# 'data': array n x 4 of 4 lengths
# 'target': array n x 1 of target classes with values 0, 1, or 2
# 'target_names': array for the 3 names of iris species values or array (['setosa', 'versicolor', 'virginica'], dtype='<U10')
# 'feature_names': list of 4 feature names.  ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
# 'frame': ...
# 'DESCR': ...
# 'filename': 'iris.csv'
# 'data_module': ...
label_names = the_data['target_names']
feature_names = the_data['feature_names']
all_labels = the_data['target']  # class label
all_features = the_data['data']  # dictionaries in python
```

# Selecting the normalization option

```python
option = '1'
# option = print(
#     'Enter 1 for normalizing the data and 0 for not normalizing the data')
# while option not in {'1', '0'}:
#     option = input(
#         'Enter 1 for normalizing the data and 0 for not normalizing the data ')
#     if option not in {'1', '0'}:
#         print("Invalid response: '%s'. Please try again." % option)
```

```python
import numpy as np
all_features_normalized = all_features
if option == '1':
    # Normalize the data(all_features) to have zero mean and unit variance, NOT using the sklearn StandardScaler

    # Compute the mean and standard deviation of each feature
    feature_means = np.mean(all_features, axis=0)
    feature_stds = np.std(all_features, axis=0)

    # Normalize the data
    all_features_normalized = (all_features - feature_means) / feature_stds
```

```python
# Looking at our data
print("\n the_data data type: ", type(the_data))
print("Feature Names: \n", feature_names)
print("\nTarget Label Names: \n", label_names)
print("\nfeatures data type: ", type(all_features))
print("\nAll Features in the data: \n", all_features)
print("\nAll Features Normalized in the data: \n", all_features_normalized)
print("\nAll Labels in the data: \n", all_labels)
```

```python
# see https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
from sklearn.model_selection import train_test_split
# Spliting our dataset into 2 parts for training and testing
training_features, testing_features, training_labels, testing_labels = train_test_split(
    all_features, all_labels, test_size=0.2, random_state=0)
print("\nTraining features: \n", training_features)
print("\nTraining Labels: \n", training_labels)
print("\nTesting Labels: \n", testing_labels)

# For normalized data
training_features_normalized, testing_features_normalized, training_labels_normalized, testing_labels_normalized = train_test_split(
    all_features_normalized, all_labels, test_size=0.2, random_state=0)
print("\nTraining features normalized: \n", training_features_normalized)
print("\nTraining Labels normalized: \n", training_labels_normalized)
print("\nTesting Labels normalized: \n", testing_labels_normalized)
```

```python
# Training our Classifier
from sklearn.naive_bayes import GaussianNB
# classifier is now an object of the Gaussian Naive Bayes class
classifier = GaussianNB()
model = classifier.fit(training_features, training_labels)
# use the model obtained in previous step to predict labels for testing features
predicted_labels = model.predict(testing_features)

# Training our Classifier for normalized data
# classifier is now an object of the Gaussian Naive Bayes class
classifier_normalized = GaussianNB()
model_normalized = classifier_normalized.fit(
    training_features_normalized, training_labels_normalized)
# use the model obtained in previous step to predict labels for testing features
predicted_labels_normalized = model_normalized.predict(
    testing_features_normalized)
```

```python
print("\nPredicted class labels: \n", predicted_labels)
print("\nCorrect Testing class labels: \n", testing_labels)
print("\nPredicted Class Names: \n", label_names[predicted_labels])
print("\nCorrect Class Names: \n", label_names[testing_labels])

# For normalized data
print("\nPredicted class labels normalized: \n", predicted_labels_normalized)
print("\nCorrect Testing class labels normalized: \n", testing_labels_normalized)
print("\nPredicted Class Names normalized: \n",
      label_names[predicted_labels_normalized])
print("\nCorrect Class Names normalized: \n",
      label_names[testing_labels_normalized])
```

```python
print("Test data where predicted label equals the test label: \n",
      testing_labels == predicted_labels)
number_correct = (testing_labels == predicted_labels).sum()
print("\nNumber of correct predictions: %d. Out of total test cases %d." %
      (number_correct, testing_labels.shape[0]))

# For normalized data
print("Test data where predicted label equals the test label normalized: \n",
      testing_labels_normalized == predicted_labels_normalized)
number_correct_normalized = (
    testing_labels_normalized == predicted_labels_normalized).sum()
print("\nNumber of correct predictions normalized: %d. Out of total test cases %d." %
      (number_correct_normalized, testing_labels_normalized.shape[0]))
```

```python
from sklearn.metrics import accuracy_score
# Calculating the % Accuracy of the prediction. For Iris dataset random_state = 0 gives 97%, random_state 40 gives 100%, random_state 5 gives 90%.
accuracy_percent = 100*accuracy_score(testing_labels, predicted_labels)
accuracy_percent_normalized = 100*accuracy_score(
    testing_labels_normalized, predicted_labels_normalized)
# %% escapes the formatting % to print '%'
print("Prediction Accuracy : %5.2f%%" % accuracy_percent)
print("Prediction Accuracy normalized : %5.2f%%" % accuracy_percent_normalized)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbor_size = []
errors_list = []
errors_list_normalized = []
for k in range(2, 20):
    classifier = KNeighborsClassifier(n_neighbors=k)
    # or can also use:   predicted_labels = classifier.fit(train_features,train_labels)
    model = classifier.fit(training_features, training_labels)
    model_normalized = classifier.fit(
        training_features_normalized, training_labels_normalized)
    # use the model obtained in previous step to predict labels for testing features
    predicted_labels = model.predict(testing_features)
    predicted_labels_normalized = model_normalized.predict(
        testing_features_normalized)
    accuracy_percent = 100*accuracy_score(testing_labels, predicted_labels)
    accuracy_percent_normalized = 100 * \
        accuracy_score(testing_labels_normalized, predicted_labels_normalized)
    # Calculating the % Accuracy of the prediction.
    print("Prediction Accuracy for k = %2d : %5.2f%%" %
          (k, accuracy_percent))  # %% escapes the formatting % to print '%'
    print("Prediction Accuracy for k = %2d normalized: %5.2f%%" % (
        k, accuracy_percent_normalized))  # %% escapes the formatting % to print '%'
    neighbor_size.append(k)
    errors_list.append(100-accuracy_percent)
    errors_list_normalized.append(100-accuracy_percent_normalized)

print(" K = ", neighbor_size, "\n", "Errors = ", errors_list)
print(" K = ", neighbor_size, "\n",
      "Errors normalized = ", errors_list_normalized)
```

```python
# plot misclassification error versus k
import matplotlib.pyplot as plt
plt.plot(neighbor_size, errors_list)  # x list and y list
plt.plot(neighbor_size, errors_list_normalized)  # x list and y list
plt.legend(['Non-Normalized', 'Normalized'])
plt.xlabel('Number of neighbors')
plt.ylabel('Misclassification Error Percent')
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
# perform k-fold cross validation
neighbor_size = [] #value of k used
errors_list = [] #accuracy for k used
errors_list_normalized = [] #accuracy for k used
num_folds = 5 #number of folds in k-fold. Usually 5 is a good number. 
for k in range (2, 20): 
    classifier = KNeighborsClassifier(n_neighbors = k)
    #try the experiment, splitting data into num_fold folds, with 1 fold as test set 
    scores = cross_val_score(classifier, all_features, all_labels, cv = num_folds, scoring = 'accuracy') 
    scores_normalized = cross_val_score(classifier, all_features_normalized, all_labels, cv = num_folds, scoring = 'accuracy')
    print("scores = ", scores, "average = ", scores.mean())
    print("scores normalized = ", scores_normalized, "average normalized = ", scores_normalized.mean())
    accuracy_percent = 100*scores.mean() #use the mean of the cv = num_folds scores as accuracy. 
    accuracy_percent_normalized = 100*scores_normalized.mean() #use the mean of the cv = num_folds scores as accuracy.
    neighbor_size.append(k)
    errors_list.append(100-accuracy_percent)
    errors_list_normalized.append(100-accuracy_percent_normalized)

print (" K = ", neighbor_size, "\n", "Error Percent = ", errors_list)
print (" K = ", neighbor_size, "\n", "Error Percent normalized = ", errors_list_normalized)
```

```python
# plot misclassification error versus k
plt.plot(neighbor_size, errors_list) #x list and y_list
# for normalized data
plt.plot(neighbor_size, errors_list_normalized) #x list and y_list
plt.xlabel('Number of neighbors (K)')
plt.ylabel('Misclassification Error Percent')
plt.legend(['Non-Normalized', 'Normalized'])
plt.grid(True)
plt.show()
```
