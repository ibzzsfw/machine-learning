```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
# Loading all features 0,1,2,3 of the iris dataset
iris = load_iris()
# features
print(f"features: {iris.feature_names}")
# name
print(f"labels: {iris.target_names}")
```

```python
# Load sepals length and petal length
X_13 = iris.data[:, [0, 2]]
y = iris.target

print(f"X_13: {X_13[0:5]}\n...")
```

```python
# Plot the data before clustering
plt.figure(figsize=(8, 6))
plt.scatter(X_13[:, 0], X_13[:, 1], c='k', s=20)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
# K-Means clustering
K_13 = 3
kmeans_13 = KMeans(n_clusters=K_13, random_state=0)
y_pred_13 = kmeans_13.fit_predict(X_13)

print(f"KMeans output: {y_pred_13}")
```

```python
# Plot
"""
class 1: <
class 2: >
class 3: ^

cluster 1: pink
cluster 2: yellow
cluster 3: cyan
"""

CLUSTER_SIZE = 40
CLASS_SIZE = 10

STYLE = {
    0: {
        'marker': '<',
        'c': 'pink',
    },
    1: {
        'marker': '>',
        'c': 'yellow',
    },
    2: {
        'marker': '^',
        'c': 'cyan',
    },
    3: {
        'marker': 'o',
        'c': 'green',
    },
}

plt.figure(figsize=(8, 6))

# cluster
for i in range(K_13):
    plt.scatter(
        X_13[y_pred_13 == i, 0],
        X_13[y_pred_13 == i, 1],
        c=STYLE[i]['c'],
        s=CLUSTER_SIZE
    )

# class
for i in range(len(iris.target_names)):
    plt.scatter(
        X_13[y == i, 0],
        X_13[y == i, 1],
        marker=STYLE[i]['marker'],
        c='k',
        s=CLASS_SIZE)

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

## 5. Redo the example in class with all 4 features and K = 3, but using your own class or function my_k_means

- In Python that has initialization parameters: K, X, max_iterations, centroid_move_epsilon and returns y as a 1-D array of integer labels of 1, 2, ..., K.. 
- Each input N-dimensional data X[i] will have a 1-dimensional output label y[i] for i = 1..M where M is the number of data points. 
- The algorithm should start by assigning K cluster centers based on random values from the (min, max) range of each dimension in the N-dimensional data X. 
- It should stop when all centers have moved by less than the centroid_move_epsilon or when the max_iterations is reached. 
- Make sure your results are similar to the K Means library class.

<image src="./algorithm.png" width=600 />

```python
class MyKMeans:
    """
    Algorithm:
    1: Specify the number k of clusters to assign.
    2: Randomly initialize k centroids. (
        The algorithm should start by assigning K cluster centers based on random values 
        from the (min, max) range of each dimension in the N-dimensional data X.
    )
    3: Repeat 
        3.1 expectation: Assign each data point to the closest centroid.
        3.2 maximization: Compute the new centroids (mean) of each cluster.
    4: Until the centroids positions do not change or a max number of iterations is reached.
    """

    def __init__(
        self,
        K: int,
        X: np.ndarray,
        max_iter: int = 100,
        centroid_move_epsilon: float = 1e-2,
        random_state: int = 0
    ):
        # 1: Specify the number k of clusters to assign.
        self.K = K
        self.X = X
        self.max_iter = max_iter
        self.centroid_move_epsilon = centroid_move_epsilon
        self.random_state = random_state

        self.labels = np.zeros(len(X))
        self.centroids = self._init_centroids()

    def fit(self):
        # 2: Randomly initialize k centroids.
        if self.centroids is None:
            self._init_centroids()

        # 4: Until the centroids positions do not change or a max number of iterations is reached.
        for _ in range(self.max_iter):
            # 3.1 expectation: Assign each data point to the closest centroid.
            self.labels = self._expectation()

            # 3.2 maximization: Compute the new centroids (mean) of each cluster.
            new_centroids = self._maximization()

            # check if the centroids have moved
            if self._check_centroid_move(new_centroids):
                break

            self.centroids = new_centroids

            # print(f"iteration {_}: {self.labels}")
            
        return self

    def _init_centroids(self):
        # The algorithm should start by assigning K cluster centers based on random values
        # from the (min, max) range of each dimension in the N-dimensional data X.
        np.random.seed(self.random_state)
        self.centroids = np.random.uniform(
            low=self.X.min(axis=0),
            high=self.X.max(axis=0),
            size=(self.K, self.X.shape[1])
        )

    def _expectation(self):
        # Assign each data point to the closest centroid.
        
        # find the distance between each data point and each centroid
        distances = np.linalg.norm(
            self.X[:, np.newaxis] - self.centroids,
            axis=2
        )

        return np.argmin(distances, axis=1)

    def _maximization(self):
        # Compute the new centroids (mean) of each cluster.
        new_centroids = np.zeros_like(self.centroids)

        for k in range(self.K):
            new_centroids[k] = np.mean(self.X[self.labels == k], axis=0)

        return new_centroids

    def _check_centroid_move(self, new_centroids):
        # check if the centroids have moved
        return np.linalg.norm(new_centroids - self.centroids) < self.centroid_move_epsilon

    def score(self, y):
        return np.sum(y == self.labels) / len(y)
```

```python
X_mine = iris.data

K_MINE = 3

kmeans_mine = MyKMeans(K=K_MINE, X=X_mine)
y_pred_mine = kmeans_mine.fit().labels

print(f"MyKMeans output: {y_pred_mine}")
print(f"Acutal labels: {y}")
```
