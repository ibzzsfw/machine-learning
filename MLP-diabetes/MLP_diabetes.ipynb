# MultiLayer Perceptron on Diabetes Dataset
Use **multilayer perceptron** on `sklearn`â€™s diabetes data meant for regression 
- using 25% test data, 
- find the R-square.

```python
# import diabetes dataset
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor

# load the diabetes dataset
diabetes = load_diabetes()
```

```python
# split the data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(
    diabetes.data,  # type: ignore
    diabetes.target,  # type: ignore
    test_size=0.25,
    random_state=0
)
```

```python
# Train the model
mlp = MLPRegressor(
    hidden_layer_sizes=(8, 8, 8),
    activation="relu",
    max_iter=1000,
)
mlp.fit(X_train, y_train)

plt.plot(mlp.loss_curve_)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.show()
```

```python
# Predict the model
predictions = mlp.predict(X_test)

# Evaluate the model
print('R-square score: %.2f' % r2_score(y_test, predictions))
```
